```python
class Question:
     def __init__(self, prompt, answer):
          self.prompt = prompt
          self.answer = answer
            
print("SPOR QUIZİ KURALLARI:")
print("1.Her sorunun bir cevabı vardır.")
print("2.Boş bırakma hakkınız vardır.")
print("3.Büyük-küçük harf farketmeksizin şıkkınızı girebilirsiniz.")
print("BAŞARILAR :)")


question_prompts = [
     "Hangi futbolcu Arjantin Milli Takımında forma giymemektedir?\n(A)Aguero\n(B)Luis Suarez\n(C)Lionel Messi\n(D)Paulo Dybala\n",
     "Aşağıdaki takımlardan hangisi Türkiye Süper Liginde şampiyon olmuştur??\n(A)Bursaspor\n(B)Ankaragücü\n(C)Eskişehirspor\n(D)Samsunspor\n",
     "Galatasaray hangi yıl UEFA kupasını almıştır?\n(A)1999\n(B)2000\n(C)2001\n(D)2002\n",
     "Cristiano Ronaldo aşağıdaki takımlardan hangisinde forma giymemiştir?\n(A)Manchester United\n(B)Sporting Lizbon\n(C)Real Madrid\n(D)Porto\n",
     "1930 yılında oynanan ilk Dünya Kupası'nı hangi ülke kazanmıştır?\n(A)Arjantin\n(B)Brezilya\n(C)Uruguay\n(D)Almanya\n",
     "Diego Maradona hangi maçta “Tanrı’nın eli” diye adlandırılan golü atmıştır?\n(A)Arjantin-Türkiye\n(B)Arjantin-Fransa\n(C)Arjantin-İngiltere\n(D)Arjantin-İtalya\n",
     "Avrupa'da maç başına ortalama taraftar sayısı en fazla olan kulüp hangisidir?\n(A)Borussia Dortmund\n(B)Bayern Munchen\n(C)Real Madrid\n(D)Manchester United\n",
     "Dünya Kupası tarihinde en çok gol atan futbolcu kimdir?\n(A)Pele\n(B)Maradona\n(C)Ronaldo\n(D)Miroslav Klose\n",
     "İstanbul'un ev sahipliği yaptığı unutulmaz Liverpool-Milan Şampiyonlar Ligi finali hangi sene oynanmıştır?\n(A)2004\n(B)2005\n(C)2006\n(D)2007\n",
     "2019 yılı şampiyonlar ligi final maçı hangi iki takım arasında oynanmıştır\n(A)Real Madrid - Barcelona\n(B)Liverpool - Barcelona\n(C)Tottenham - Bayern Munich\n(D)Tottenham - Liverpool\n",
     
]

questions = [
     Question(question_prompts[0], "B"),
     Question(question_prompts[1], "A"),
     Question(question_prompts[2], "B"),
     Question(question_prompts[3], "D"),
     Question(question_prompts[4], "C"),
     Question(question_prompts[5], "C"),
     Question(question_prompts[6], "A"),
     Question(question_prompts[7], "D"),
     Question(question_prompts[8], "B"),
     Question(question_prompts[9], "D"),
]

def run_quiz(questions):
    score = 0
    empty = 0
        
    for question in questions:
        answer = input(question.prompt)
        answer = answer.upper()
          
        
        if answer == "":
            empty += 1
               
          
        elif answer == question.answer:
                  score += 10
                    
    if empty > 5:
        print("5 soru veya daha az soru cevapladınız. BAŞARISIZ!!!")
    
    elif empty < 6 and score < 51:
        print("100 üzerinden", score, "aldınız. BAŞARISIZ!!!")
    
    elif empty < 6 and score > 51:
        print("100 üzerinden", score, "aldınız. BAŞARILI!!!")
              
    

run_quiz(questions)
```
